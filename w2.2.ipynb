```python
#set
#no duplicates, no index (no order), iterable (can loop through), collection, mutable (can change).
set1 = {"Ben", "Jason", "Darlene", "Jason", "Erin"}
#basically never do it this way
set_declared = set(("Ben", "Jason", "G,", "G", "Darlene"))

print (f"set1: {set1}")
print (f"set_declared: {set_declared}")
```

```python
#method to find overlaps
overlaps = set1.intersection(set_declared)
overlaps
```

```python
#Method to find differnece between two sets
diff = set1.difference(set_declared)
diff
```

```python
set1.add("Halle")
set1
```

```python
set1.remove("Darlene")
set1
```

```python
#tuple: a collection that exists in parentheses
#non-mutable (cannot change), dublicates, iterable, indexed
tuple1 = (1,2,3,4)
tuple_declared = tuple((3,4,5,6))

print(f"tuple1: {tuple1}")
print(f"tuple_declared: {tuple_declared}")
```

```python
tuple1 = (1,2,3,4,5) #overwriting not changing
tuple1
```

```python
tuple1.count(2)
#not many methods because it is non-mutable
```

```python
tuple1.index(1)
#returns the index of the value given
```

```python
print(f"tuple1 = {tuple1}")
tuple1[0]
#returns value at index 0
```

```python
#list
#mutable, indexed, iterable, duplicates
list1 = ["G", "G", "Paola", "Emil"] #could make empty becuase can always add later
list_declared = list(("G", "G", "Paola", "Emil"))

print(f"list1 = {list1}")
print(f"list_declared = {list_declared}")
```

```python
#index notation [:]

print(f"list1 = {list1}")
print(f"first index = {list1[0]}")
print(f"last item in list1 = {list1[-1]}") #start at -1 for last, then -2 for second to last, etc.
print(f"series between 2 & 3 items = {list1[1:3]}") #start at index 1, stop before index 3 (python is exclusive)
print(f"everything before index 3 = {list1[:3]}")
print(f"everything after the first = {list1[1:]}") #first item is inclusive
```

```python
#list methods
list1.append("Clare") # add to a list, wont return a new list
list2 = list1.append("Nathan")
print(f"appended Clare = {list1}")
print(f" createdlist2 = {list2}") #None because append does not return anything
removed = list1.pop() # remove
print(f"Removed {removed}. New list1 is {list1}") 
list1[1] = "J" # change index
print(f"Index 1 of list1 changed. New list1 = {list1}")
combined_list = [list1, list_declared] # combine lists
print(f"combined_list = {combined_list}")
combined_lists = [*list1, *list_declared] # combine lists into one list
print(f"combining lists into 1: {combined_lists}")
list1.extend(list_declared)# combine lists
print(f"list1 after extension = {list1}")
```
