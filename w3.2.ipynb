```python
#if statements

condition = ""

if condition:
   # pass #does nothing lets me continue logic wihtout error
    print("Condition is True")
elif bool(condition) == False: #this evaluates condition logic, wrap in bool to convert to True or False, wont reach else now
    #pass
    print("Condition is False")
else:
    print("I got to the else statement")
```

```python
#hwo we should actually do if statement
#introducing not condition

condition = ""

if condition:
   # pass #does nothing lets me continue logic wihtout error
    print("Condition is True")
elif not condition: 
    #pass
    print("Condition is False")
else:
    print("I got to the else statement")

print(f"not statement = {not condition}")
```

```python
num = input("Please enter a number between 1 and 10")
num = int(num)

if num < 3:
    print(f"num is less than 3: {num}")
elif num > 6:
    print(f"num is greater than 6: {num}")
else:
    print(f"num is between 3 and 6: {num}")
```

```python
# lets make this better wiht try except

num = input("Please enter a number between 1 and 10")

try:
    num = int(num)

    if num < 3:
        print(f"num is less than 3: {num}")
    elif num > 6:
        print(f"num is greater than 6: {num}")
    else:
        print(f"num is between 3 and 6: {num}")
except Exception as e:
    print(f"Well that didnt work, try agian: {e}")
```

```python
#type hinting: str, int, float, bool
#nested if statements

variable: str = input(f"Please, enter a value")

if variable.isnumeric():
    try:
        num: float = float(variable)
        if num > 10:
            print(f"The value is too high, please start over")
        else:
            print(f"{num} is a good choice")
    except Exception as e:
        print(f"There was an error: {e}")
elif variable.isalpha():
    print(f"You chose: {variable.upper()}")
else:
    print("Please enter either a number or letter")
```
