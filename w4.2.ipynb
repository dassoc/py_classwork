```python
#while: does something WHILE a condition is true, creates infinity loop
count = 0
while True:
    if count >= 10:
        break #breaks the loop
    count += 1 #count=count+1
    print(count)
```

```python
#while loops: a better option
count = 0

while count < 10:
    count += 1
    print(count)
```

```python
#while example
is_pass_complete: bool = False

while not is_pass_complete: 
    password = input("Enter password in only alphabetic characters")
    if not password.isalpha():
        print("Please chose another passoword with only alphabetic characters")
        continue #continues the loop from here. reads nothing below
    is_pass_complete = True
    (print("Password set"))
    
```

```python
#for loop
from typing import List
nums: List[int] = [1, 2, 3, 4, 5]

for num in nums: 
#num is temporary variable used to call each item in collection in code block
    num +=1
    if num % 3 == 0:
       #  break would break the loop
       continue #do not proint anything divisible by 3
    print(num)
```

```python
#nested loops

sentence = "This is a Python course."

words:List[str] = sentence.split(" ")
used_vowels = []

for word in words:
    vowels = ["a", "e", "i", "o", "u", "y"]
    for letter in word:
        if letter in vowels:
            used_vowels.append(letter)
        

print(used_vowels)
```

```python
#list comprehension/generative loop/one line loop

bar_names = List[str] = ["Ottoman Empire", "France", "Spain"]

bar_names = [name[:3] + "." for name in bar_names]

print(bar_names)

#for name in bar_names:
 #   print(name{:4})
```

```python
# what are functions
#function take an input (argument), processes it
# return an output (return value)

def shorten_names(names_list: List[str], chars: int = 3) -> List[str]: #arguments go in parentheses
    """
    returns a list of strings that are th firs tthree characters of the word wiht a period appended.
    chars (int): the number of characters to shorten each string to. must be greate than o, defaults to 3
    Args: 
        names_list (List[str]): a list of strings to shorten
    Returns:
        List[str]: a list of shortened string to three characters.
    """
    # """ blah """ is a documetnation string, not necessary but nice to have
    if type(chars) != int:
        raise TypeError("chars must be an int")
    if chars < 0:
        raise ValueError("chars must be greater than 0")
    names_list = [f"{name[:chars]}." for name in names_list]
    return names_list
```

```python
bar_names: List[str] = ["Ottoman Empire", "France", "Spain"]
shorten_names(bar_names, chars=4)
```
