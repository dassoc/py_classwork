## univariable visualization

```python
import pandas as pd
import matplotlib.pyplot as plt
#%conda install seaborn #ifseaborn not installed
import seaborn as sns
```

```python
url = "https://raw.githubusercontent.com/nkmwicz/data-for-students/refs/heads/main/soccer-players.csv"

df = pd.read_csv(url)
df.head
```

```python
#anatomy of a viz, never set plt equal to a variable
plt.figure(figsize=(5,3))
plt.xlabel("Height in CM")
plt.ylabel("Frequency")
plt.title("Histogram of Heights (cm) of Soccer Players")
plt.xlim(0,210)
plt.ylim(0,500)
plt.grid(True)
text = "skew: 0.12"
#plt.text(x=110, y=250, s=text, ha="center", va="center", fontsize=12, color="blue") dont like this so 
#doing it differnet way
plt.text(
    x=0.05, 
    y=0.95, 
    s=text, 
    #ha="center", 
    va="center", 
    fontsize=12, 
    #color="blue",
    transform=plt.gca().transAxes)
plt.axvline(x=105, color="r", linestyle="--", label="Mean")
plt.axhline(y=250, color="b", linestyle="--", label="Center")
plt.legend()
plt.show()
```

```python
import scipy.stats as stats
```

```python
#histogram in plt
plt.figure(figsize=(5,3))
hts = df["height_in_cm"]
bins = round(hts.count() ** (1/3))
plt.hist(x=hts, bins=bins, edgecolor="black")
plt.xlabel("Height in CM")
plt.ylabel("Frequency")
plt.title("Histogram of Heights (cm) of Soccer Players")
plt.grid(True)
sk = stats.skew(hts, nan_policy="omit")
kurt = stats.kurtosis(hts, nan_policy="omit")
text = f"Skew: {sk:.3f}\nKurtosis: {kurt:.3f}"
#plt.text(x=110, y=250, s=text, ha="center", va="center", fontsize=12, color="blue") dont like this so 
#doing it differnet way
plt.text(
    x=0.05, 
    y=0.90, 
    s=text, 
    #ha="center", 
    va="center", 
    fontsize=12, 
    #color="blue",
    transform=plt.gca().transAxes)
plt.axvline(x=hts.mean(), color="r", linestyle="--", label="Mean")
plt.axvline(x=hts.median(), color="g", linestyle="--", label="Median")
plt.legend()
plt.show()
```
